```python
import pandas as pd
import os
import numpy as np
```

```python
PATH = r'/Users/jan/Documents/working papers/project 1/data/btc_1h.csv'

df = pd.read_csv(PATH)

df["timestamp"] = pd.to_datetime(df["timestamp"])
```

```python
def trim_df(df, start_year, start_month, n_rows):
    # Filter by year and month
    df_filtered = df[
        (df["timestamp"].dt.year >= start_year)
    ]
    
    # Take the first n_rows
    return df_filtered.head(n_rows)
```

```python

df_trimmed = trim_df(df, start_year=2018, start_month=6, n_rows=10000)
```

```python
print(df_trimmed)
```

```python
save_dir = os.path.join(os.getcwd(), "/Users/jan/Documents/working papers/project 1/data")
os.makedirs(save_dir, exist_ok=True)  

df_trimmed.to_csv(os.path.join(save_dir, "btc_1h_test.csv"), index=False)
```
